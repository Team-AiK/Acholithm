```python
import numpy as np
import re
import sys
from operator import eq


str1 = input()

if len(str1) < 2 or len(str1) > 1000:
    print('2자 이상 1000자 이하로 입력하세요')
    sys.exit(1)

str2 = input()

if len(str2) < 2 or len(str2) > 1000:
    print('2자 이상 1000자 이하로 입력하세요')
    sys.exit(1)

str1 = str1.lower()
str2 = str2.lower()
str1_list = []
str2_list = []


for i in range(len(str1) - 1):
    if eq(re.sub('[^가-힝a-zA-Z]','', str1[i:i+2]) , str1[i:i+2] ):
        str1_list.append(str1[i:i+2])

for i in range(len(str2) - 1):
    if eq(re.sub('[^가-힝a-zA-Z]','', str2[i:i+2]) , str2[i:i+2] ):
        str2_list.append(str2[i:i+2])


union = np.union1d(str1_list,str2_list)
intersect = np.intersect1d(str1_list,str2_list)


def multiset_intersection(str1_list, str2_list, intersect):
    count_1 = 0
    count_2 = 0
    intersect_range = len(intersect)


    for i in range(intersect_range):
        for j in range(len(str1_list)):
            if eq(str1_list[j],intersect[i]):
                count_1 = count_1 + 1

        for j in range(len(str2_list)):
            if eq(str2_list[j],intersect[i]):
                count_2 = count_2 + 1

        if count_1 > 1 or count_2 > 1:
            for j in range(min(count_1, count_2) - 1):

                intersect = np.append(intersect,intersect[i])

        count_1=0
        count_2=0


    return intersect

def multiset_union(str1_list, str2_list, union):
    count_1 = 0
    count_2 = 0
    union_range = len(union)

    for i in range(union_range):
        for j in range(len(str1_list)):
            if eq(str1_list[j],union[i]):
                count_1 = count_1 + 1

        for j in range(len(str2_list)):
            if eq(str2_list[j],union[i]):
                count_2 = count_2 + 1

        if count_1 > 1 or count_2 > 1:
            for j in range(max(count_1, count_2) - 1):
                union = np.append(union,union[i])

        count_1=0
        count_2=0
        
    return union

intersect = multiset_intersection(str1_list, str2_list, intersect)
union = multiset_union(str1_list, str2_list, union)

if union.size == 0:
    print('65536')
else:
    print(int(len(intersect)/len(union) * 65536))
```
